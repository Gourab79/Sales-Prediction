```python
import pandas as pd
df =pd.read_excel('/content/Advertising.xlsx')
# Printing dataset
df
```

```python
# Assigning independent vaiables as x and dependent variables as y
x= df[['TV','Radio','Newspaper']]
y = df['Sales']
```

```python
# Creating train test data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

```python
# Creating linear model for prediction test data set
from sklearn.linear_model import LinearRegression
reg =LinearRegression()
```

```python
# Fitting training dataset
reg.fit(x_train,y_train)
```

```python
# prediction process
reg.predict(x_test)
```

```python
# testing scores of prediction
reg.score(x_test,y_test)
```
